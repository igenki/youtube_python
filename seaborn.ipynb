```python
# !pip install seaborn
import seaborn as sns
```

## seabornで使えるデータの表示

```python
sns.get_dataset_names()
```

## チップのデータの読み込みと表示

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

## 散布図

```python
sns.scatterplot(x="tip", y="total_bill", data=tips)
```

```python
# 曜日ごとに表示を分ける
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day")
```

```python
# さらにサイズごとに分ける
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day", size="size")
```

```python
# カラーパレットの変更
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day", size="size", palette="YlGnBu")
```

## ヒストグラム

```python
sns.histplot(tips['tip'], kde=True, bins=15)
```

## 棒グラフ

```python
sns.barplot(x="sex", y="tip", data=tips, palette="GnBu")
```

## 箱ひげ図

```python
sns.boxplot(x="day", y="tip", data=tips, hue="sex", palette="YlGnBu")
```

## バイオリン図

```python
sns.violinplot(x="day", y="tip", data=tips, hue="sex", palette="YlGnBu")
```

## ストリッププロット

```python
sns.stripplot(x="day", y="tip", data=tips, hue="sex", palette="YlGnBu", dodge=True)
```

## 散布図＋ヒストグラム

```python
sns.jointplot(x="tip", y="total_bill", data=tips)
```

```python
# 回帰直線の追加
sns.jointplot(x="tip", y="total_bill", data=tips, kind="reg")
```

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind="hex")
```

## ペアプロット図（散布図行列）

```python
sns.pairplot(data=tips)
```

## ヒートマップ

```python
tips.corr()
```

```python
sns.heatmap(tips.corr(), annot=True, cmap="YlGnBu")
```
