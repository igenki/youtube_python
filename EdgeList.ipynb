## 無向グラフ

```python
%matplotlib inline
# NetworkXをインポート
import networkx as nx

# ネットワーク生成
# グラフデータの読み込み
G = nx.read_edgelist('r_edgelist.txt')
nx.draw(G, with_labels=True) # ラベルをTrueにして番号の可視化

# グラフデータの書き出し
nx.write_edgelist(G, 'w_edgelist.txt')
```

## 重み付き無向グラフ

```python
# 重み付きグラフデータの読み込み
G = nx.read_edgelist('r_weighted_edgelist.txt')

# レイアウト
pos = nx.spring_layout(G)

# 重みの表示設定
edge_labels=dict([((u,v,),d['weight']) for u,v,d in G.edges(data=True)])
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# 辺の色のリスト作成
edge_color = [edge['color'] for edge in G.edges.values()]

# 描画
nx.draw(G, pos, node_color='red', edge_color=edge_color, with_labels=True) # ラベルをTrueにして番号の可視化

# 重み付きグラフデータの書き出し
nx.write_edgelist(G, 'w_weighted_edgelist.txt')
```

## 重み付き有向グラフ

```python
# 重み付きグラフデータの読み込み
G = nx.read_edgelist('r_weighted_edgelist.txt', create_using=nx.DiGraph())

# レイアウト
pos = nx.circular_layout(G)

# 重みの表示設定
edge_labels=dict([((u,v,),d['weight']) for u,v,d in G.edges(data=True)])
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# 辺の色のリスト作成
edge_color = [edge['color'] for edge in G.edges.values()]

# 描画
nx.draw(G, pos, node_color='green', edge_color=edge_color, with_labels=True) # ラベルをTrueにして番号の可視化

# 重み付きグラフデータの書き出し
nx.write_edgelist(G, 'w_weighted_edgelist.txt')
```
