## BAネットワーク

```python
%matplotlib inline
import networkx as nx
import EoN # EoNのインポート
import matplotlib.pyplot as plt # Matplotlibのインポート

N = 10**5 # 個体数
G=nx.barabasi_albert_graph(N, 5) # BAネットワークの作成

# 感染に関するパラメータ
gamma = 1.0
tau = 0.3
rho = 0.005

# GillespieアルゴリズムによるSIRのシミュレーション実行
t, S, I, R = EoN.Gillespie_SIR(G, tau, gamma,rho = rho)

# 個体数変化のグラフ
plt.plot(t, S, color='green', label='S')
plt.plot(t, I, color='red', label='I')
plt.plot(t, R, color='blue', label='R')
plt.xlabel('t')
plt.legend()
```

## ERネットワーク

```python
%matplotlib inline
import networkx as nx
import EoN # EoNのインポート
import matplotlib.pyplot as plt # Matplotlibのインポート

N = 10**4 # 個体数
G=nx.erdos_renyi_graph(N, 0.1) # ERネットワークの作成

# 感染に関するパラメータ
gamma = 1.0
tau = 0.3
rho = 0.005

# GillespieアルゴリズムによるSIRのシミュレーション実行
t, S, I, R = EoN.Gillespie_SIR(G, tau, gamma,rho = rho)

# 個体数変化のグラフ
plt.plot(t, S, color='green', label='S')
plt.plot(t, I, color='red', label='I')
plt.plot(t, R, color='blue', label='R')
plt.xlabel('t')
plt.legend()
```
