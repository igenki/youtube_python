```python
%matplotlib inline
import networkx as nx # NetworkXをインポート
import matplotlib.pyplot as plt
import numpy as np

# ネットワーク生成
G = nx.Graph([(1, 2), (2, 3), (3, 1)])
# G = nx.Graph() # 空のグラフを作成する場合
nx.draw(G, with_labels=True) # ラベルをTrueにして番号の可視化
```

```python
# ノードの追加
G.add_node(4)
nx.draw(G, with_labels=True)
```

```python
# リンクの追加
G.add_edge(1,4)
nx.draw(G, with_labels=True)
```

```python
# 複数ノードの追加
n = [5, 6, 7]
G.add_nodes_from(n)

# 複数リンクの追加
e = [(5,6), (3,7), (4,6)]
G.add_edges_from(e)

nx.draw(G, with_labels=True)
```

```python
# ノードの削除
G.remove_node(4)
nx.draw(G, with_labels=True)
```

```python
# 複数リンクの削除
G.remove_edges_from([(1,2), (2,3)])
nx.draw(G, with_labels=True)
```
