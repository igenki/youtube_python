```python
# 散布図

# Jupyterで画像表示のため以下の1行が必要
%matplotlib inline
 
import numpy as np
import matplotlib.pyplot as plt
 
# 乱数を生成
x = np.random.rand(10)
y = np.random.rand(10)

print(x,y)
 
# 散布図を描画
plt.scatter(x, y)
#plt.savefig("scatter.png")
plt.show()
```

```python
# グラフのタイトル，x軸，y軸の名前 (ラベル)，グリッド線を表示

plt.scatter(x, y)
plt.title('散布図', fontname="MS Gothic")
plt.xlabel('x軸', fontname="MS Gothic")
plt.ylabel('y軸', fontname="MS Gothic")
plt.grid(True)
```

```python
# 折れ線グラフ

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
 
# 折れ線グラフを出力
x = [1, 2, 3, 4, 5]
#x = np.array([1, 2, 3, 4, 5])
y = np.random.rand(5)
plt.plot(x, y)
```

```python
# 複数の折れ線グラフ
plt.plot(x, y, linewidth=2)
plt.plot(x, y/2, linewidth=2, linestyle='dashed')
```
