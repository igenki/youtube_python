# 様々なネットワークの次数分布

対応する動画：[【研究で使うPython】#12 NetworkXによるネットワーク分析　⑤次数と次数分布](https://youtu.be/czTzLQ0DUyM)

動画のプレイリスト：[研究で使うPython](https://www.youtube.com/playlist?list=PLw5AEs-T7pLHxLaNvGD9tnhrDxAgH-QIw)

```python
%matplotlib inline
import networkx as nx # NetworkXをインポート
import matplotlib.pyplot as plt
import numpy as np

N = 10 # ノード数

# ランダムグラフ生成
rnd = nx.gnp_random_graph(N, 0.1)
pos=nx.circular_layout(rnd)

# 特徴量出力
print(nx.info(rnd))
print('density:', nx.density(rnd), ', clustering coefficient:', nx.average_clustering(rnd))

# グラフ描画
plt.subplot(1,2,1)
plt.title('random graph')
nx.draw(rnd, pos, with_labels=True)

# 次数分布
plt.subplot(1,2,2)
degree_dist = [i/N for i in nx.degree_histogram(rnd)]
plt.plot(degree_dist)
plt.xlabel('$k$')
plt.ylabel('$p(k)$')
plt.ylim(0,1)
```

```python
# サイクルグラフ生成
cycle = nx.cycle_graph(N)
pos=nx.spring_layout(cycle)

# 特徴量出力
print(nx.info(cycle))
print('density:', nx.density(cycle), ', clustering coefficient:', nx.average_clustering(cycle))

# グラフ描画
plt.subplot(1,2,1)
plt.title('cycle graph')
nx.draw(cycle, pos, with_labels=True)

# 次数分布
plt.subplot(1,2,2)
degree_dist = [i/N for i in nx.degree_histogram(cycle)]
plt.plot(degree_dist)
plt.xlabel('$k$')
plt.ylabel('$p(k)$')
plt.ylim(0,1)
```

```python
# 完全グラフ生成
K_10 = nx.complete_graph(N)
pos=nx.spring_layout(K_10)

# 特徴量出力
print(nx.info(K_10))
print('density:', nx.density(K_10), ', clustering coefficient:', nx.average_clustering(K_10))

# グラフ描画
plt.subplot(1,2,1)
plt.title('complete graph')
nx.draw(K_10, pos, with_labels=True)

# 次数分布
plt.subplot(1,2,2)
degree_dist = [i/N for i in nx.degree_histogram(K_10)]
plt.plot(degree_dist)
plt.xlabel('$k$')
plt.ylabel('$p(k)$')
plt.ylim(0,1)
```
