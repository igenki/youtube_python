## ランダムグラフの生成と密度とクラスタ係数の計算

```python
import networkx as nx
import numpy as np

# ノード数
num_nodes = 10
# リンクを張る確率
probability = 0.1

# Erdős-Rényiランダムグラフの生成
G = nx.erdos_renyi_graph(num_nodes, probability)

# ランダムグラフの密度を計算
density = nx.density(G)

# ランダムグラフの平均クラスタ係数を計算
average_clustering = nx.average_clustering(G)

# 結果の出力
print("Erdős-Rényiランダムグラフの密度:", density)
print("Erdős-Rényiランダムグラフの平均クラスタ係数:", average_clustering)
```

### ネットワークの可視化の追加

```python
import networkx as nx
import matplotlib.pyplot as plt

# ノード数
num_nodes = 10
# リンクを張る確率
probability = 0.1

# Erdős-Rényiランダムグラフの生成
G = nx.erdos_renyi_graph(num_nodes, probability)

# ランダムグラフの密度を計算
density = nx.density(G)

# ランダムグラフの平均クラスタ係数を計算
average_clustering = nx.average_clustering(G)

# グラフの可視化
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=300, node_color='skyblue', font_size=10)
plt.title("Erdős-Rényiランダムグラフ", fontname="MS Gothic")
plt.show()

# 結果の出力
print("Erdős-Rényiランダムグラフの密度:", density)
print("Erdős-Rényiランダムグラフの平均クラスタ係数:", average_clustering)
```

## 完全グラフの生成と密度とクラスタ係数の計算

```python
import networkx as nx
import matplotlib.pyplot as plt

# ノード数
num_nodes = 10

# 完全グラフの生成
G = nx.complete_graph(num_nodes)

# グラフの可視化
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=300, node_color='skyblue', font_size=10)
plt.title("完全グラフ",  fontname="MS Gothic")
plt.show()

# ランダムグラフの密度を計算
density = nx.density(G)

# ランダムグラフの平均クラスタ係数を計算
average_clustering = nx.average_clustering(G)

# 結果の出力
print("完全グラフの密度:", density)
print("完全グラフの平均クラスタ係数:", average_clustering)
```

## ランダムグラフの生成と次数分布の計算

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# ノード数
num_nodes = 10
# リンクを張る確率
probability = 0.1

# ランダムグラフの生成
G = nx.erdos_renyi_graph(num_nodes, probability)

# グラフの可視化
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=300, node_color='skyblue', font_size=10)
plt.title("ランダムグラフ", fontname="MS Gothic")
plt.show()

# 次数分布を計算
degrees = [G.degree(node) for node in G.nodes()]
degree_values, degree_counts = np.unique(degrees, return_counts=True)

# 次数分布をプロット
plt.bar(degree_values, degree_counts, width=0.5, color='skyblue')
plt.xlabel('次数', fontname="MS Gothic")
plt.ylabel('ノード数', fontname="MS Gothic")
plt.title('次数分布', fontname="MS Gothic")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# ノード数
num_nodes = 10

# スターグラフの生成
G = nx.star_graph(num_nodes - 1)

# グラフの可視化
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=300, node_color='skyblue', font_size=10)
plt.title("スターグラフ", fontname="MS Gothic")
plt.show()

# 次数分布を計算
degrees = [G.degree(node) for node in G.nodes()]
degree_values, degree_counts = np.unique(degrees, return_counts=True)

# 次数分布の割合を計算
total_nodes = len(G.nodes())
degree_percentages = degree_counts / total_nodes

# 次数分布をプロット
plt.bar(degree_values, degree_percentages, width=0.5, color='skyblue')
plt.xlabel('次数', fontname="MS Gothic")
plt.ylabel('割合', fontname="MS Gothic")
plt.title('次数分布', fontname="MS Gothic")
plt.show()
```

## 距離（最短経路長）を計算するためのネットワーク作成

```python
import networkx as nx
import matplotlib.pyplot as plt

# グラフの作成
G = nx.Graph()

# エッジの追加
edges = [(1, 2), (1, 3), (1, 4), (1, 5), (3, 4), (4, 5), (4, 6), (5, 6)]
G.add_edges_from(edges)

# グラフの可視化
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=300, node_color='skyblue', font_size=10, font_color='black', node_shape='o')
plt.title("指定されたネットワーク", fontname="MS Gothic")
plt.show()
```

### ノード間の距離計算

```python
import networkx as nx

# グラフの作成
G = nx.Graph()

# エッジの追加
edges = [(1, 2), (1, 3), (1, 4), (1, 5), (3, 4), (4, 5), (4, 6), (5, 6)]
G.add_edges_from(edges)

# ノード1とノード2の最短経路長を計算
distance_1_to_2 = nx.shortest_path_length(G, source=1, target=2)

# ノード3とノード6の最短経路長を計算
distance_3_to_6 = nx.shortest_path_length(G, source=3, target=6)

# 結果の出力
print("ノード1とノード2の間の距離:", distance_1_to_2)
print("ノード3とノード6の間の最短経路長:", distance_3_to_6)
```

### 平均距離（平均最短経路長）と直径

```python
import networkx as nx

# グラフの作成
G = nx.Graph()

# エッジの追加
edges = [(1, 2), (1, 3), (1, 4), (1, 5), (3, 4), (4, 5), (4, 6), (5, 6)]
G.add_edges_from(edges)

# 平均距離を計算
average_distance = nx.average_shortest_path_length(G)

# 直径を計算
diameter = nx.diameter(G)

# 結果の出力
print("ネットワークの平均距離:", average_distance)
print("ネットワークの直径:", diameter)
```
