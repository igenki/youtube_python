```python
import networkx as nx
from networkx.drawing.nx_agraph import graphviz_layout
```

```python
# 階層的なネットワークの生成
G = nx.DiGraph([(1,2),(1,3),
                (2,4),(2,5),(3,6),(3,7),
                (4,8),(4,9),(5,10),(5,11),(6,12),(6,13),(7,14),(7,15)])
```

```python
# NetworkXで描画
nx.draw(G, with_labels=True)
```

```python
# Graphvizで描画
pos = graphviz_layout(G, prog="dot") # 階層型の"dot"というレイアウトを使用
nx.draw(G, pos=pos, with_labels=True) # pos=posでGraphvizのレイアウト通りのグラフを描画
```

```python
pos = graphviz_layout(G,prog="neato") # ばねモデルの"neato"というレイアウトを使用
nx.draw(G, pos=pos, with_labels=True) 
```

```python
pos = graphviz_layout(G, prog="twopi") # 放射状に配置する"twopi"というレイアウトを使用
nx.draw(G, pos=pos, with_labels=True) 
```

```python
pos = graphviz_layout(G, prog="circo") # 環状に配置する"twopi"というレイアウトを使用
nx.draw(G, pos=pos, with_labels=True) 
```

```python
# サイズが30のスターグラフをNetworkXで生成
G = nx.star_graph(30)
```

```python
pos = graphviz_layout(G,prog="dot")
nx.draw(G,pos=pos,with_labels=True) 
```

```python
pos = graphviz_layout(G,prog="neato")
nx.draw(G,pos=pos,with_labels=True) 
```

```python
pos = graphviz_layout(G,prog="twopi")
nx.draw(G,pos=pos,with_labels=True) 
```

```python
pos = graphviz_layout(G,prog="circo")
nx.draw(G,pos=pos,with_labels=True) 
```
