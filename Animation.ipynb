```python
# sin波のアニメーション

# nbaggはJupyter上でインタラクティブな画像表示を実現するための機能
%matplotlib nbagg

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

x = np.arange(0, 10, 0.1)

# Figureのインスタンス生成
fig = plt.figure()

# 画像データをストックするためのリスト
ims = []

for a in range(50):
    y = np.sin(x - a)
    
    # 画像データの生成
    im = plt.plot(x, y, 'r')
    
    # 画像データをリストに追加
    # scatterやimshowのようにデータ型がリスト出ない場合は，[im]のようにリストにラッピングしてから渡す
    ims.append(im)

ani = animation.ArtistAnimation(fig, ims)
plt.show()
```
