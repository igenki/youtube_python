```python
%matplotlib inline
import networkx as nx # NetworkXをインポート

# ネットワーク生成
G = nx.Graph([(1, 2), (2, 3), (3, 1)])
# G = nx.Graph() # 空のグラフを作成する場合
nx.draw(G, with_labels=True) # ラベルをTrueにして番号の可視化
```

```python
# ノードの追加
G.add_node(4)
nx.draw(G, with_labels=True)
```

```python
# リンクの追加
G.add_edge(1,4)
nx.draw(G, with_labels=True)
```

```python
# 複数ノードの追加
n = [5, 6, 7]
G.add_nodes_from(n)

# 複数リンクの追加
e = [(5,6), (3,7), (4,6)]
G.add_edges_from(e)

nx.draw(G, with_labels=True)
```

# ランダムグラフの作成

```python
N = 50 # ノード数

# ランダムグラフ生成
rnd = nx.gnp_random_graph(N, 0.1)
pos=nx.circular_layout(rnd)
#pos=nx.spring_layout(rnd)

# 特徴量出力
print(nx.info(rnd))

# グラフ描画
nx.draw(rnd, pos)
```
