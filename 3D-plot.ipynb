# 3次元のグラフ

対応する動画：[Pythonで3次元のグラフとアニメーションを作成する【研究で使うPython #28】](https://www.youtube.com/watch?v=G3aPIHwiCsw)

```python
%matplotlib inline
import numpy as np
import random
import matplotlib.pyplot as plt

def func(x, y):
    return x**2 + y**2

fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

x = np.arange(-3.0, 3.1, 0.1)
y = np.arange(-3.0, 3.1, 0.1)
# 格子点の生成
X, Y = np.meshgrid(x, y)
#print(X)
#print(Y)

ax.plot_wireframe(X, Y, func(X, Y), color='b',rstride=2, cstride=2, linewidth=0.3)
#ax.plot_surface(X, Y, func(X, Y))

# -3から3までの乱数のリストを生成
x_data = 6 * np.random.rand(50) - 3
y_data = 6 * np.random.rand(50) - 3
#print(x_data)

ax.scatter3D(x_data, y_data, func(x_data, y_data), c='r' )
```
